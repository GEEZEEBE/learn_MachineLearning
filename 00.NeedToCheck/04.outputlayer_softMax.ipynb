# SoftMax(주로 Classication 사용)

```
import numpy as np
```

```

def softmax(a):
  norm1 = a / np.linalg.norm(a)
  exp_a = np.exp(a)
  sum_exp_a = np.sum(exp_a)
  y = exp_a / sum_exp_a
  print("지수 함수에 대입하여 나온 수치:", exp_a)
  print("모든 지수의 합", sum_exp_a)
  print("소프트맥스 값", y)
  return y
```

```
a = np.array([0.9, 3.7, 4.0])
softmax(a)
```
