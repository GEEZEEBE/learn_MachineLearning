## SSE(Error Sum of Squares) - 선형 경우

```
import numpy as np
import matplotlib.pyplot as plt
```

```
def sse(predict_y, true_y):
  return np.sum(np.square(np.array(true_y) - np.array(predict_y))) / 2
```

```
y = [0.05, 0.1, 0.0, 0.6, 0.06, 0.1, 0.0, 0.0, 0.0]
_y = [0,1,0,1,0,0,1,0,0]
```

```
E = sse(y,_y)
E
```

## CEE(Cross Entopy Error) - 회귀 경우

```
%matplotlib inline
x = np.arange(0, 1, 0.001)
y = np.log(x)
plt.plot(x,y)
```

```
def cee(predict_y, true_y):
  delta = 1e-10 # 무한 0 발생 시 보존값
  E = - np.sum(true_y * np.log(predict_y + delta))
  return E
  
  
```

```
y = np.array([0.1, 0.05, 0.05, 0, 0.2, 0.5, 0, 0, 0.05])
_y = np.array([0,0,0,0,0,0,1,0,0])
E = cee(_y, y)
E
```

##GradientDescent

```
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(x):
  return np.square(x)

def f_bias(x):
  bias = -4
  return 4 * x + bias
```

```
x = np.arange(-10, 10, 0.1)
```

```
plt.plot(x, f(x))
plt.plot(x, f_bias(x))
```

```
def numerical_diff(f,x):
  h = 1e-10
  return(f(x+h) - f(x-h)) / (2*h)
```

```
plt.plot(numerical_diff(f,x))
```

### Exam(뱡향은 ?,에러값은  Weights)

```
def sigmoid(x):
  return 1 / (1+np.exp(x))

def sigmoid_prime(x):
  return sigmoid(x) * (1-sigmoid(x))
```

```
input1 = np.array([0.5, -0.3])
```

```
weights = np.array([0.4,0.6])
```

```
y = 0.5
```

```
learning_rate = 0.1
```

```
h = np.dot(input1, weights)
h
```

```
output = sigmoid(h)
output
```

```
error = y - output
error_term = error * sigmoid_prime(h)
delta_w = learning_rate * error_term * input1
```

```
print(error, error_term, delta_w)
```
