```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

```python
pd_series = pd.DataFrame(np.random.randn(100,5).cumsum(0), columns = [1,2,3,4,5])
```

```python
pd_series.plot()
```

```python
pd_series.plot(kind = 'bar')
```

```python
pd_series.plot(kind = 'barh')
```

```python
pd_series.plot(kind = 'barh', stacked=True)
```

```python
pd_series.hist(bins=50)
```

```python
pd_series.hist(bins=10, normed=True)
pd_series.plot(kind='kde')
```
