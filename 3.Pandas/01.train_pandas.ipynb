```python
import pandas as pd

dan1 = pd.Series(range(10))
dan2 = dan1 * 2
dan3 = dan1 * 3
data = {
    'first': dan1,
    'second': dan2,
    'third': dan3
}
```

```python
pd_df = pd.DataFrame(data)
pd_df
```

```python
pd_df.T
```

```python
pd_df.index = ['영', '일', '이', '영', '일', '이', '영', '일', '이', '열']
pd_df
```

```python
import numpy as np
dan4 = np.arange(10) * 4
pd_df['4단'] = dan4
pd_df
```

```python
dan11 = np.arange(10) * 11
pd_df['four'] = dan11
pd_df
```

```python
pd_df.columns = ['11단', '2단', '3단', '4단', '5', '6']
pd_df
```
