```

import numpy as np

from numpy import mean, cov
from numpy.linalg import eig
# define a matrix
# A = array([[1, 2], [3, 4], [5, 6]])
arr = np.array([[95, 95], [90, 95], [80, 75], [60, 70], [40, 35], [80, 80], [95, 90], [30, 25], [15, 10], [60, 70]])
# print('input Data matrix :', arr)
# calculate the mean of each column
# M = mean(A.T, axis=1)
# print('calculate the mean of each column : ', M)
# center columns by subtracting column means
# C = A - M

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
stand_arr = scaler.fit_transform(arr)

print(stand_arr.mean().round(), stand_arr.var())

print('StandardScaler matrix : ',stand_arr)
```

```
# calculate covariance matrix of centered matrix
covarMatrix = cov(stand_arr.T, bias=True)
print('covariance matrix : ', covarMatrix)
# eigendecomposition of covariance matrix
eigvalues, eigvectors = eig(covarMatrix)
print('eigen values : ', eigvalues)
print('eigen vetors : ', eigvectors)
# project data
# P = vectors.T.dot(C.T)
# print(P.T)
pca_data = eigvectors.round().T.dot(stand_arr.T)
print(pca_data.T)
```

```
# np.corrcoef(stand_arr.T, pca_data)
```

```
from scipy.stats.stats import pearsonr   
pearsonr(np.round(pca_data.T[:,0]), pca_data.T[:,1])
```

```
np.correlate(stand_arr.T, pca_data)
```

```
from sklearn.decomposition import PCA
pca = PCA()
# fit on data
# pca.fit(stand_arr)
# access values and vectors
# print(pca.components_)
# transform data
pca_result = pca.fit_transform(stand_arr)
print(pca.components_)
print(pca.explained_variance_)
pca_result.var(), pca_result.mean() , pca_result
```
