#statsmodels.api with OLS

```
import statsmodels.api as sm
import numpy as np
all_data = [
 [0,2,6],
 [1,3,9],
 [2,4,11.5],
 [3,5,14.5],
 ] 
ndarray = np.array(all_data)
x_data = ndarray[:,:-1]
y_data = ndarray[:,-1]
x_data, y_data
```

```
sm_constant = sm.add_constant(x_data)
estimate = sm.OLS(y_data, sm_constant)
results = estimate.fit()
print(results.summary())
```

# R-squared

```
from sklearn import metrics
import numpy as np

# 예시
predict_data = [1, 2, 3, 4, 5, 7, 9, 10]
true_data = [1, 10, 3, 3, 9, 7, 8, 10]

print('MAE:',metrics.mean_absolute_error(true_data, predict_data))  # 기본
print('MSE:',metrics.mean_squared_error(true_data, predict_data))
print('RMSE:', np.sqrt(metrics.mean_squared_error(true_data, predict_data)))
print('R2 Score:',metrics.r2_score(true_data, predict_data)) # 1에 가까울 수록 좋음.
```
